```python
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
import json
import os
```

```python
model_type="bert"
model_name= "bert-base-cased"
```

```python
root_dir = "/content/drive/MyDrive/data/effective_date_parser"
```

```python
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "use_cached_eval_features": True,
    "output_dir": os.path.join(root_dir, model_type),
    "best_model_dir":os.path.join(root_dir, model_type, 'best_model'),
    "evaluate_during_training": True,
    "num_train_epochs": 5,
    "save_model_every_epoch": True,
    "save_eval_checkpoints": False,
    "train_batch_size": 16,
    "eval_batch_size": 4,
    "wandb_project" : "Effective Date Parser",
    "wandb_kwargs" : {"name" : model_name},
    "use_early_stopping" : True
}
```

```python
model = QuestionAnsweringModel(
    model_type,model_name, args=train_args
)
```

```python
with open("/content/drive/MyDrive/data/effective_date_parser/processed_train.json", 'r') as f:
  train = json.load(f)
```

```python
test = train[-50:]
final_train = train[:-50]
```

```python
model.train_model(final_train, eval_data=test)
```

<h2>Evaluating trained model</h2>

```python
with open("/content/drive/MyDrive/data/effective_date_parser/processed_eval.json", 'r') as f:
  eval_data = json.load(f)
```

```python
result, text = model.eval_model(eval_data)
```

<h1> Testing prediction file</h1>

```python
with open(os.path.join(root_dir, 'processed_test.json'), 'r') as f:
  test_data = json.load(f)
```

```python
best_model = os.path.join(root_dir, model_type, 'best_model')
model = QuestionAnsweringModel(
    model_type,best_model
)
```

```python
answers, probabilities = model.predict(test_data)

print(answers)
```
